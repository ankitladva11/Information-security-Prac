## vignere_cipher

```python
import math
```

```python
pt="harekrishana"
key="ankit"
```

```python
def generate_key(key,pt):
    n=math.ceil(len(pt)/len(key))
    key="".join([key]*n)
    key=key[:len(pt)]
    return key
    
key=generate_key(key,pt)
```

```python
pt=pt.upper()
key=key.upper()
ct=[]
for i in range(len(pt)):
    a=ord(key[i])+ord(pt[i])
    a=a%26
    a+=ord("A")
    ct.append(chr(a))
ct
        
```

```python
# pt=pt.upper()
# key=key.upper()
ppt=[]
for i in range(len(ct)):
    a=ord(ct[i])-ord(key[i])+26
    a=a%26
    a+=ord("A")
    ppt.append(chr(a))
ppt
        
```
